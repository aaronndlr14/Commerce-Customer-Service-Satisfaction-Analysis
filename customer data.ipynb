```python
agent_scores = df.groupby('agent_name')['csat_score'].agg(['mean', 'count']).reset_index()
agent_scores = agent_scores[agent_scores['count']>=50]
top10 = agent_scores.sort_values('mean', ascending=False).head(10)
bottom10 = agent_scores.sort_values('mean').head(10)

combined = pd.concat([top10, bottom10])


combined = combined.sort_values('mean', ascending= True)
```

```python
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(10,8))


sns.barplot(data=combined, x='mean', y='agent_name')


plt.title('Top 10 and Bottom 10 Agents by Average CSAT score')


plt.xlabel('Average CSAT Score')

plt.ylabel('Agent Name')

plt.legend()

plt.tight_layout()

plt.axvline(x=df['csat_score'].mean(), color='black', linestyle='--', label='Overall Average')


plt.show()
```

```python
category_scores = df.groupby('category')['csat_score'].mean().sort_values(ascending=False).reset_index()


category_scores.columns = ['category', 'mean_csat']

plt.figure(figsize= (10,8))


sns.barplot(data=category_scores, x = 'mean_csat', y= 'category')

plt.title('Average CSAT Score by Category')

plt.xlabel('Average CSAT Score')

plt.ylabel('Category')

plt.axvline(x=df['csat_score'].mean(), color='black', linestyle='--', label='Overall Average')

plt.legend()

plt.tight_layout()

plt.show()
```

```python
shift_scores = df.groupby('agent_shift')['csat_score'].mean().sort_values(ascending = False).reset_index()


shift_scores.columns = ['agent_shift', 'mean_csat']

plt.figure(figsize=(10,8))

sns.barplot(data=shift_scores, x = 'mean_csat', y= 'agent_shift')

plt.title('Average CSAT Score by Agent Shift')

plt.xlabel=('Average CSAT Score')

plt.ylabel=('Agent Shift')

plt.axvline(x=df['csat_score'].mean(), color='black', linestyle='--', label='Overall Average')

plt.legend()

plt.tight_layout()

plt.show()
```

```python
df=df.rename(columns={'sub-category': 'sub_category'})
```

```python
subcat_scores = df[df['category'].isin(['Returns', "Order Related"])].groupby(['category', 'sub_category'])['csat_score'].agg(['mean', 'count']).reset_index()


subcat_scores = subcat_scores[subcat_scores['count']>=50]


subcat_scores= subcat_scores.sort_values('mean', ascending=True)

plt.figure(figsize=(12,8))


sns.barplot(data=subcat_scores, x= 'mean', y= 'sub_category')

plt.title('Average CSAT Score by Sub-Category (Returns & Order Related)')

plt.xlabel('Average CSAT Score')

plt.ylabel('Sub Category')

plt.axvline(x=df['csat_score'].mean(), color='black', linestyle='--', label='Overall Average')

plt.legend(title='Category')

plt.tight_layout()

plt.show()

del xlabel
```
